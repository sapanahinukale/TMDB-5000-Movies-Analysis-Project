```python
# Display title of the movie having runtime > 40 
tmdb_data = tmdb_data[tmdb_data['runtime']>40]
tmdb_data
```

```python
# Plot a histogram of movie runtime
plt.hist(tmdb_data['runtime'],range=(40,200), bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Runtimes')
plt.show()
```

```python
# Visualize the distribution of original languages
language_counts = tmdb_data['original_title'].value_counts().head(15)
sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')
plt.xlabel('Original title')
plt.ylabel('Count')
plt.title('Top 10 Original titles in Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot a scatter plot of budget vs. revenue
plt.scatter(tmdb_data['budget'], tmdb_data['revenue'], alpha=0.5)
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget vs. Revenue')
plt.show()
```

```python
# Plot a line plot of average revenue over release years
avg_revenue_by_year = tmdb_data.groupby(tmdb_data['release_date'].dt.year)['revenue'].mean()
avg_revenue_by_year.plot(marker='o', linestyle='-')
plt.xlabel('Release Year')
plt.ylabel('Average Revenue')
plt.title('Average Revenue Over Release Years')
plt.show()
```
