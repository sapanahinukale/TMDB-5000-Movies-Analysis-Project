# Visualization

```python
# Plot a histogram of movie runtime
plt.hist(tmdb_data['runtime'], bins=15, edgecolor='black')  # Create a histogram with 15 bins and black edges
plt.xlabel('Runtime')  # Label the x-axis as 'Runtime'
plt.ylabel('Frequency')  # Label the y-axis as 'Frequency'
plt.title('Distribution of Movie Runtimes')  # Set the title of the plot
plt.show()  # Display the plot
```

This graph provides insight into the distribution of movie runtimes within the dataset, showing the most common durations and the overall spread of movie lengths.

```python
# Visualize the distribution of original titles in movies
# Count the occurrences of each original title and select the top 10
language_counts = tmdb_data['runtime'].value_counts().head(10)

# Create a bar plot with seaborn
sns.barplot(x=language_counts.index, y=language_counts.values, palette='viridis')

# Label the x-axis as 'runtime' and the y-axis as 'Count'
plt.xlabel('runtime')
plt.ylabel('Count')

# Set the title of the plot
plt.title('Top 10 Original titles in Movies')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

This graph provides a visual representation of the distribution of movie runtimes for the top 10 original titles, allowing for easy comparison of runtime frequencies across different titles. 

It counts the occurrences of each movie runtime value and selects the top 10 most common runtimes

```python
# Plotting a scatter plot of budget vs. revenue using Seaborn
sns.scatterplot(x='budget', y='revenue',data=tmdb_data)  # Scatter plot using Seaborn

# Adding labels and title to the plot
plt.xlabel('Budget')  # Label for the x-axis as 'budget'
plt.ylabel('Revenue')  # Label for the y-axis as 'revenue'
plt.title('Budget vs. Revenue')  # Title of the plot

# Displaying the plot
plt.show()  # Show the plot
```

This graph provides insights into the relationship between movie budget and revenue, highlighting the general trend as well as the variability in revenue across different budget levels.

```python
# Plot a line plot of average revenue over release years
# Grouping the data by release year and calculating the average revenue
avg_revenue_by_year = tmdb_data.groupby(tmdb_data['release_date'].dt.year)['revenue'].mean()

# Plotting the line plot
plt.plot(avg_revenue_by_year.index, avg_revenue_by_year.values, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Release Year')
plt.ylabel('Average Revenue')
plt.title('Average Revenue Over Release Years')

# Displaying the plot
plt.show()
```

 This graph provides insights into the average revenue trends in the movie industry over the years, highlighting both the general trend of increasing revenue and the year-to-year variability.
